```python
# Simulating Prover's pre-preparation of quantum states based on guessed bases and bits
```

```python
n = 10000

b_bases = np.random.randint(2, size=n)
eve_bases = np.random.randint(2, size=n)

# Create random bits n long
eve_bits = np.random.randint(2, size=n)
c_bits = np.random.randint(2, size=n)
```

```python
print("Eve's bits: ", eve_bits)
print("Eve's bases: ", eve_bases)
print("Bob's bases: ", b_bases)
print("C bits: ", c_bits)
```

```python
# Simulating Prover's pre-preparation of quantum states based on guessed bases and bits
message = encode_message(eve_bits, eve_bases, n)
```

```python
# Verifier decoding the message using their actual bases
c_prime = decode_message(message, b_bases, n)
```

```python
# Verifier compares c and c'
print("\nc bits:", c_bits)
print("\nc' bits:", c_prime)
print(np.array_equal(c_bits, c_prime))

counter = 0
for i in range(n):
    if c_bits[i] == c_prime[i]:
        counter += 1
success_rate = counter / n
print(f"\nAdversary success rate: {success_rate:.6f}")
```
