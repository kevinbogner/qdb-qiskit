```python
import numpy as np
from rotation_functions import *

%reload_ext jupyter_black
```

```python
n = 10000
otp_length = 3 * n
one_time_pad = np.random.randint(2, size=otp_length)  # Generate the one-time pad

# Extract a_bases, b_bases, and c_bits from the one-time pad
a_bases = one_time_pad[:n]
b_bases = one_time_pad[n : 2 * n]
c_bits = one_time_pad[2 * n : 3 * n]

# Print extracted values for verification
print("a bases:", a_bases)
print("b bases:", b_bases)
print("c bits:", c_bits)
```

```python
# Encode c in a
message = encode_message(c_bits, a_bases, n)

# Draw all the circuits after encoding on Verifier's side
# for i in range(n):
#    display(message[i].draw(output="mpl", style="iqp"))
```

```python
# Prover compares a and b bases and applies a rotation
message_back = apply_specific_rotation(a_bases, b_bases, message)

# Draw all the circuits after the rotation
# for i in range(n):
#    display(message_back[i].draw(output="mpl", style="iqp"))
```

```python
# Verifier reverts the rotation and measures the qubits
c_prime = decode_message(message_back, b_bases, n)
```

```python
# Verifier compares c and c'
print("\nc bits:", c_bits)
print("\nc' bits:", c_prime)
print(np.array_equal(c_bits, c_prime))
```
